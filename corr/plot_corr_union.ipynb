```python
def plot(x, y, str1,str2,r):
    import numpy as np
    import matplotlib.pyplot as plt
    import matplotlib as mpl
    mpl.rcParams['pdf.fonttype'] = 42
    # Initialize layout
    fig, ax = plt.subplots(figsize = (8,8))

    # Add scatterplot
    ax.scatter(np.log2(x), np.log2(y), alpha=0.1,s=40,rasterized=True)
    plt.title("union")
    my_x_ticks = np.arange(0, 10, 2)
    my_y_ticks = np.arange(0, 10, 2)
    plt.xticks(my_x_ticks)
    plt.yticks(my_y_ticks)
    plt.xlabel(f"{str1} raw data")
    plt.ylabel(f"{str2} raw data")
    plt.xlim(0,10)
    plt.ylim(0,10)
    plt.text(1,9,format(r, '.2f'),fontsize=16, color ="red",fontweight = "bold")
    plt.savefig(f"{str1}_{str2}.pdf")
    plt.show()

def quantile_normalize(df):
    import pandas as pd
    import numpy as np
    """
    input: dataframe with numerical columns
    output: dataframe with quantile normalized values
    """
    df_sorted = pd.DataFrame(np.sort(df.values,
                                     axis=0), 
                             index=df.index, 
                             columns=df.columns)
    df_mean = df_sorted.mean(axis=1)
    df_mean.index = np.arange(1, len(df_mean) + 1)
    df_qn =df.rank(method="min").stack().astype(int).map(df_mean).unstack()
    return(df_qn)

def corr(file,str1,str2):
    import pandas as pd
    import scipy.stats
    df = pd.read_csv(file, sep = "\t", header = None)
    df.columns = [str1,str2]
    r, p = scipy.stats.pearsonr(df[str1],df[str2])
    df1 = quantile_normalize(df)
    plot(df1[str1],df1[str2],str1,str2,r)
```
